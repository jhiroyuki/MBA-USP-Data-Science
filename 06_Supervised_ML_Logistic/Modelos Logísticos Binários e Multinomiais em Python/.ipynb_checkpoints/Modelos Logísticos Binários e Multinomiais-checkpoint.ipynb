### - Sklearn

Os resultados encontrados aqui são diferentes do passado, provavelmente porque o sklearn usa um implementação diferente. Note que os valores encontrados são próximos, não havendo grandes diferenças, portanto é uma questão de aprofundamento da teoria para que se possa decidir com mais clareza qual biblioteca oferece os resultados mais coerentes.

```python
from sklearn.linear_model import LogisticRegression

# Estimação de modelo logístico binário
X = df[['dist', 'sem']]
y = df[['atrasado']]

llm = LogisticRegression(penalty='none',  solver='newton-cg' ) #binary logistic model
llm = llm.fit(X.values, y.values)

# Não há tabela de resultados como o statsmodel. 

# Vamos criar uma função para tirar os resultados dos coeficientes e realizar os testes
from scipy import stats
def getParams_llm(X, y, llm):

    index = ['(Intercept)'] + list(X.columns.values)
    params = np.append(llm.intercept_,llm.coef_)
    df = pd.DataFrame({'Coefficients': params}, index = index)
    std_error = stderror_llm(X, y, llm)
    z_values, p_values = z_test_llm(X, y, llm)
    df = pd.DataFrame({'Coefficients': params, "Standard Errors": std_error,
                          "z-statistics": z_values, 'p-values': p_values}, index = index)
    other_df = OtherTests_llm(X, y, llm)
    return df, other_df


def stderror_llm(X, y, llm):
    if isinstance(X, pd.DataFrame):
        if len(X.columns)==1:
            X = X.values.reshape(-1,1)
            y = y.values
        else:
            X = X.values
            y = y.values
    
    if llm.intercept_ ==0:
        k = len(X[0,:] ) # Graus de liberade de regressão (graus de inclinação apenas)
        n = len(X[:, 0]) # Numero de observações
    else:
        k = len(X[0,:] ) + 1 # Graus de liberade de regressão (graus de inclinação + intercepto)
        n = len(X[:, 0]) # Numero de observações

    params = np.append(llm.intercept_,llm.coef_)
    predictions = llm.predict(X) 
    
    
    X_matrix = np.insert(X, 0, 1, axis = 1) # Matriz de design
    V_matrix = np.diag(llm.predict_proba(X)[:, 0]*llm.predict_proba(X)[:, 1])
    C_matrix = np.linalg.inv(np.matmul(X_matrix.T,np.matmul(V_matrix, X_matrix))) # Variance-covariance matrix
    C_diag = np.diag(C_matrix)
    std_error = np.sqrt(C_diag) # erros padrões
    
    return std_error

# Teste z
def z_test_llm(X, y, llm):
    
    if isinstance(X, pd.DataFrame):
        if len(X.columns)==1:
            X = X.values.reshape(-1,1)
            y = y.values
        else:
            X = X.values
            y = y.values

    
    if llm.intercept_ ==0:
        k = len(X[0,:] ) # Graus de liberade de regressão (graus de inclinação apenas)
        n = len(X[:, 0]) # Numero de observações
    else:
        k = len(X[0,:] ) + 1 # Graus de liberade de regressão (graus de inclinação + intercepto)
        n = len(X[:, 0]) # Numero de observações
    
    params = np.append(llm.intercept_,llm.coef_)
    std_error = stderror_llm(X, y, llm)
    z_values = params/std_error # Coeficientes do teste t para os parâmetros do modelo
    p_values =[2*(1-stats.norm.cdf(abs(i))) for i in z_values]
    
    
    return z_values, p_values

def OtherTests_llm(X, y, llm):
    
    if isinstance(X, pd.DataFrame):
        if len(X.columns)==1:
            X = X.values.reshape(-1,1)
            y = y.values
        else:
            X = X.values
            y = y.values
    
    if llm.intercept_ ==0:
        k = len(X[0,:] ) # Graus de liberade de regressão (graus de inclinação apenas)
        n = len(X[:, 0]) # Numero de observações
    else:
        k = len(X[0,:] ) + 1 # Graus de liberade de regressão (graus de inclinação + intercepto)
        n = len(X[:, 0]) # Numero de observações
        
    # Modelo nulo
    X_null = np.ones_like(y).reshape(-1, 1)
    llm_null = LogisticRegression(penalty='none',  solver='newton-cg' ) #binary logistic model
    llm_null = llm_null.fit(X_null, y)
    
    # Loglikelihood
    loglike = np.sum(np.ravel(y) * np.log(llm.predict_proba(X)[:, 1]) + (1- np.ravel(y))* np.log(llm.predict_proba(X)[:, 0]))
    loglike0 = np.sum(np.ravel(y) * np.log(llm_null.predict_proba(X_null)[:, 1]) + (1- np.ravel(y))* np.log(llm_null.predict_proba(X_null)[:, 0]))
    # Teste Chi squared
    chi2_value = -2*(loglike0 - loglike)
    p_value = 1-stats.chi2.cdf(chi2_value, k)

    
    # Demais parâmetros
    AIC = -2*loglike + 2*k 
    BIC = -2*loglike + k*np.log(n)  
    pR2_McFadden = chi2_value/(-2*loglike0)
    pR2_Cragg = (1 - (np.exp(loglike0)/np.exp(loglike))**(2/n))/(1 - np.exp(loglike0)**(2/n))
    
    
    pd.reset_option('display.float_format')
    values = [int(n),n -k, len(X[0,:] ),  chi2_value, p_value, loglike, AIC, BIC, pR2_McFadden, pR2_Cragg]
    index = ['No. Observations', 'Df Residuals', 'Df Model','Pearson chi2', 'p-value (Pearson chi2)', 'Log-Likelihood', 'AIC', 'BIC', 'pseudo-R2 (McFadden)', 'pseudo-R2 (Cragg)']
    df = pd.DataFrame({'Values': values}, index = index)

    return df
    

df_params, df_other = getParams_llm(X, y, llm)
```

```python
def confint_llm(X, y, llm, significance =0.05):
    index = ['(Intercept)'] + list(X.columns.values)
    
    if isinstance(X, pd.DataFrame):
        if len(X.columns)==1:
            X = X.values.reshape(-1,1)
            y = y.values
        else:
            X = X.values
            y = y.values
    
    if llm.intercept_ ==0:
        k = len(X[0,:] ) # Graus de liberade de regressão (graus de inclinação apenas)
        n = len(X[:, 0]) # Numero de observações
    else:
        k = len(X[0,:] ) + 1 # Graus de liberade de regressão (graus de inclinação + intercepto)
        n = len(X[:, 0]) # Numero de observações

    params = np.append(llm.intercept_,llm.coef_)
    
    stderror = stderror_llm(X, y, llm) 
    z_critical = np.abs(stats.norm.ppf(significance/2)) # Valor em x da distriuição normal que corresponde ao valor de significancia
    
    params_correction = z_critical*stderror
    params_low = params - params_correction
    params_high = params + params_correction

    df = pd.DataFrame({'Coefficients': params,str(significance/2*100) +'%': params_low,str((1-significance/2)*100) +'%': params_high }, index= index)
    return df

confint_llm(X, y, llm, significance =0.05)
```

## 1.5. Matriz de confusão

```python
# In[ ]: Construção de função para a definição da matriz de confusão

from sklearn.metrics import confusion_matrix, accuracy_score,\
    ConfusionMatrixDisplay, recall_score

def matriz_confusao(observado,predicts,cutoff):
    
    values = predicts.values
    
    predicao_binaria = []
    
    for item in values:
        if item >= cutoff:
            predicao_binaria.append(1)
        else:
            predicao_binaria.append(0)
    
    cm = confusion_matrix(observado, predicao_binaria)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
    plt.show()
    
    sensitividade = recall_score(observado, predicao_binaria, pos_label=1)
    especificidadee = recall_score(observado, predicao_binaria, pos_label=0)
    acuracia = accuracy_score(observado, predicao_binaria)

    #Visualizando os principais indicadores desta matriz de confusão
    indicadores = pd.DataFrame({'Sensitividade':[sensitividade],
                                'Especificidade':[especificidadee],
                                'Acurácia':[acuracia]})
    return indicadores
```

```python
#Matriz de confusão para cutoff = 0.5
matriz_confusao(observado=df['atrasado'],
                predicts=df['phat'], 
                cutoff=0.5)
```

### - Igualando critérios de especificidade e de sensitividade

```python
# Igualando critérios de especificidade e de sensitividade

#Tentaremos estabelecer um critério que iguale a probabilidade de
#acerto daqueles que chegarão atrasados (sensitividade) e a probabilidade de
#acerto daqueles que não chegarão atrasados (especificidade).

#ATENÇÃO: o que será feito a seguir possui fins didáticos, apenas. DE NENHUMA
#FORMA o procedimento garante a maximização da acurácia do modelo!

#Criação da função 'espec_sens' para a construção de um dataset com diferentes
#valores de cutoff, sensitividade e especificidade:

def espec_sens(observado,predicts):
    
    # adicionar objeto com os valores dos predicts
    values = predicts.values
    
    # range dos cutoffs a serem analisados em steps de 0.01
    cutoffs = np.arange(0,1.01,0.01)
    
    # Listas que receberão os resultados de especificidade e sensitividade
    lista_sensitividade = []
    lista_especificidade = []
    
    for cutoff in cutoffs:
        
        predicao_binaria = []
        
        # Definindo resultado binário de acordo com o predict
        for item in values:
            if item >= cutoff:
                predicao_binaria.append(1)
            else:
                predicao_binaria.append(0)
                
        # Cálculo da sensitividade e especificidade no cutoff
        sensitividade = recall_score(observado, predicao_binaria, pos_label=1)
        especificidadee = recall_score(observado, predicao_binaria, pos_label=0)
        
        # Adicionar valores nas listas
        lista_sensitividade.append(sensitividade)
        lista_especificidade.append(especificidadee)
        
    # Criar dataframe com os resultados nos seus respectivos cutoffs
    resultado = pd.DataFrame({'cutoffs':cutoffs,'sensitividade':lista_sensitividade,'especificidade':lista_especificidade})
    return resultado

```

```python
# Até o momento, foram extraídos 3 vetores: 'sensitividade',
#'especificidade' e 'cutoffs'. Assim, criamos um dataframe que contém
#os vetores mencionados

dados_plotagem = espec_sens(observado = df['atrasado'],
                            predicts = df['phat'])
dados_plotagem
```

```python
# Visualizando o novo dataframe 'dados_plotagem' e plotando os dados
#em um gráfico que mostra a variação da especificidade e da sensitividade
#em função do cutoff

plt.figure(figsize=(10,10))
plt.plot(dados_plotagem.cutoffs,dados_plotagem.sensitividade, '-o',
         color="#440154FF")
plt.plot(dados_plotagem.cutoffs,dados_plotagem.especificidade, '-o',
         color="#2ecc71")
plt.legend(['Sensitividade', 'Especificidade'], fontsize=17)
plt.xlabel("Cuttoff", fontsize=14)
plt.ylabel("Sensitividade / Especificidade", fontsize=14)
plt.show()

```

```python
# In[ ]: Construção de função para a definição da matriz de confusão

from sklearn.metrics import confusion_matrix, accuracy_score,\
    ConfusionMatrixDisplay, recall_score

def matriz_confusao(observado,predicts,cutoff):
    
    values = predicts.values
    
    predicao_binaria = []
    
    for item in values:
        if item >= cutoff:
            predicao_binaria.append(1)
        else:
            predicao_binaria.append(0)
    
    cm = confusion_matrix(observado, predicao_binaria)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
    plt.show()
    
    sensitividade = recall_score(observado, predicao_binaria, pos_label=1)
    especificidadee = recall_score(observado, predicao_binaria, pos_label=0)
    acuracia = accuracy_score(observado, predicao_binaria)

    #Visualizando os principais indicadores desta matriz de confusão
    indicadores = pd.DataFrame({'Sensitividade':[sensitividade],
                                'Especificidade':[especificidadee],
                                'Acurácia':[acuracia]})
    return indicadores
```

## 4.2. Estimação de modelo 

### - Statsmodel

```python
# In[ ]: Estimação do modelo logístico multinomial

from statsmodels.discrete.discrete_model import MNLogit

x = df.drop(columns=['estudante','atrasado','atrasado2'])
y = df['atrasado2']

#Esse pacote precisa que a constante seja definida pelo usuário
X = sm.add_constant(x)

#Estimação do modelo - função 'MNLogit' do pacote
#'statsmodels.discrete.discrete_model'
modelo = MNLogit(endog=y, exog=X).fit()

#Parâmetros do modelo
modelo.summary()
```

### - Sklearn

```python
from sklearn.linear_model import LogisticRegression

X = df[['dist', 'sem']]
y = df[['atrasado2']]

mllm = LogisticRegression(penalty='none',  solver='newton-cg', multi_class='multinomial' ) #binary logistic model
mllm = mllm.fit(X.values,y)

# Vamos criar uma função para tirar os resultados dos coeficientes e realizar os testes
from scipy import stats
def getParams_mllm(X, y, mllm):
    
    index = ['(Intercept)'] + list(X.columns.values)
    
    alfa = np.delete(mllm.intercept_, 0) - mllm.intercept_[0]
    beta = np.delete(mllm.coef_, 0, axis = 0) - mllm.coef_[0, :]
    
    categories_model = np.delete(np.unique(y.values), 0)
    df_list = []
    
    std_error_list = stderror_mllm(X, y, mllm)
    z_values_list, p_values_list = z_test_mllm(X, y, mllm)
    for i in range(len(alfa)):
        params = np.append(alfa[i],beta[i,:])
        std_error = std_error_list[i]
        z_values = z_values_list[i]
        p_values = p_values_list[i]
        
        df_aux = pd.DataFrame({'Coefficients': params, "Standard Errors": std_error,  "z-statistics": z_values, 'p-values': p_values}, index = index)
        df_aux.columns.name = y.columns.values[0] + '=' + str(categories_model[i])
        df_list.append(df_aux)
    
    
    other_df = OtherTests_mllm(X, y, mllm)
    return df_list, other_df


def stderror_mllm(X, y, mllm):
    if isinstance(X, pd.DataFrame):
        if len(X.columns)==1:
            X = X.values.reshape(-1,1)
            y = y.values
        else:
            X = X.values
            y = y.values
    
 
    k = len(X[0,:] ) + 1 # Graus de liberade de regressão (graus de inclinação + intercepto)
    n = len(X[:, 0]) # Numero de observações
    
    categories_model = np.delete(np.unique(y), 0)
    
    list_I = []
    X_matrix = np.insert(X, 0, 1, axis = 1) # Matriz de design
    for i in categories_model:
        row_I = []
        for j in categories_model:
           
            if i==j:
                V_matrix = np.diag((mllm.predict_proba(X)[:, i])*(1 - mllm.predict_proba(X)[:, i]))
                I_matrix = np.matmul(X_matrix.T,np.matmul(V_matrix, X_matrix)) # Variance-covariance matrix
            else:
                V_matrix = np.diag((mllm.predict_proba(X)[:, j])*(mllm.predict_proba(X)[:, i]))
                I_matrix = -np.matmul(X_matrix.T,np.matmul(V_matrix, X_matrix)) # Variance-covariance matrix
            row_I.append(I_matrix)
        list_I.append(row_I)
    

    I_matrix = np.block(list_I)
    C_matrix = np.linalg.inv(I_matrix)
    C_diag = np.diag(C_matrix)
    std_error = np.sqrt(C_diag)
    std_error = std_error.reshape( len(categories_model), k)
    
    return std_error


def z_test_mllm(X, y, mllm):
    
    if isinstance(X, pd.DataFrame):
        if len(X.columns)==1:
            X = X.values.reshape(-1,1)
            y = y.values
        else:
            X = X.values
            y = y.values

    

    k = len(X[0,:] ) + 1 # Graus de liberade de regressão (graus de inclinação + intercepto)
    n = len(X[:, 0]) # Numero de observações
    
    
    alfa = np.delete(mllm.intercept_, 0) - mllm.intercept_[0]
    beta = np.delete(mllm.coef_, 0, axis = 0) - mllm.coef_[0, :]
    
    
    std_error_list = stderror_mllm(X, y, mllm)
    
    list_z = []
    list_pvalues = []
    for i in range(len(alfa)):
        params = np.append(alfa[i],beta[i,:])
        std_error = std_error_list[i]
        
        z_values = params/std_error # Coeficientes do teste t para os parâmetros do modelo
        p_values =[2*(1-stats.norm.cdf(abs(i))) for i in z_values]
        
        list_z.append(z_values)
        list_pvalues.append(p_values)
    
    
    return list_z, list_pvalues



def OtherTests_mllm(X, y, mllm):
    dummies = pd.get_dummies(y, columns = y.columns.values).values
    
    if isinstance(X, pd.DataFrame):
        if len(X.columns)==1:
            X = X.values.reshape(-1,1)
            y = y.values
        else:
            X = X.values
            y = y.values
    

    k = len(X[0,:] ) + 1  # Graus de liberade de regressão (graus de inclinação + intercepto)
    n = len(X[:, 0]) # Numero de observações
        
    # Modelo nulo
    X_null = np.ones_like(y).reshape(-1, 1)
    mllm_null = LogisticRegression(penalty='none',  solver='newton-cg', multi_class='multinomial' ) 
    mllm_null = mllm_null.fit(X_null, y)
    
    # Loglikel
    
    loglike = (dummies *np.log(mllm.predict_proba(X))).sum()
    loglike0 = (dummies *np.log(mllm_null.predict_proba(X_null))).sum()
    # Teste Chi squared
    chi2_value = -2*(loglike0 - loglike)
    p_value = 1-stats.chi2.cdf(chi2_value, k)
    # Demais parâmetros
    AIC = -2*loglike + 4*k 
    BIC = -2*loglike + 2*k*np.log(n)  
    pR2_McFadden = chi2_value/(-2*loglike0)
    pR2_Cragg = (1 - (np.exp(loglike0)/np.exp(loglike))**(2/n))/(1 - np.exp(loglike0)**(2/n))
    
    
    pd.reset_option('display.float_format')
    values = [int(n),n -2*k, 2*len(X[0,:] ),  chi2_value, p_value, loglike, AIC, BIC, pR2_McFadden, pR2_Cragg]
    index = ['No. Observations', 'Df Residuals', 'Df Model','Chi2', 'p-value (Chi2)', 'Log-Likelihood', 'AIC', 'BIC', 'pseudo-R2 (McFadden)', 'pseudo-R2 (Cragg)']
    #values = [int(n),n -k, len(X[0,:] ),  loglike, AIC, BIC]
    #index = ['No. Observations', 'Df Residuals', 'Df Model', 'Log-Likelihood', 'AIC', 'BIC']
    
    df = pd.DataFrame({'Values': values}, index = index)

    return df

df_params, other_df = getParams_mllm(X, y, mllm)
```
