### - sklearn.linear_model

```python
#Estimação do modelo
from sklearn.linear_model import PoissonRegressor
X = df[['staff', 'post', 'corruption']]
y = df[['violations']]


plm = PoissonRegressor(alpha = 0, solver='newton-cholesky')  #Poisson Linear Model
plm = plm.fit(X.values, np.ravel(y.values))


def getParams_plm(X, y, plm):

    index = ['(Intercept)'] + list(X.columns.values)
    params = np.append(plm.intercept_,plm.coef_)
    std_error = stderror_plm(X, y, plm)
    z_values, p_values = z_test_plm(X, y, plm)
    df = pd.DataFrame({'Coefficients': params, "Standard Errors": std_error, "z-statistics": z_values, 'p-values': p_values}, index = index)
    
    other_df = OtherTests_plm(X, y, plm)
    return df, other_df
    


def stderror_plm(X, y, plm):
    if isinstance(X, pd.DataFrame):
        if len(X.columns)==1:
            X = X.values.reshape(-1,1)
            y = y.values
        else:
            X = X.values
            y = y.values
    
    if plm.intercept_ ==0:
        k = len(X[0,:] ) # Graus de liberade de regressão (graus de inclinação apenas)
        n = len(X[:, 0]) # Numero de observações
    else:
        k = len(X[0,:] ) + 1 # Graus de liberade de regressão (graus de inclinação + intercepto)
        n = len(X[:, 0]) # Numero de observações

    params = np.append(plm.intercept_,plm.coef_)
    
    
    X_matrix = np.insert(X, 0, 1, axis = 1)
    W_matrix = np.diag(plm.predict(X))
    C_matrix = np.linalg.inv(np.matmul(X_matrix.T,np.matmul(W_matrix, X_matrix)))# Variance-covariance matrix
    C_diag = np.diag(C_matrix)
    std_error = np.sqrt(C_diag)
    
    return std_error


def z_test_plm(X, y, plm):
    
    if isinstance(X, pd.DataFrame):
        if len(X.columns)==1:
            X = X.values.reshape(-1,1)
            y = y.values
        else:
            X = X.values
            y = y.values

    
    if plm.intercept_ ==0:
        k = len(X[0,:] ) # Graus de liberade de regressão (graus de inclinação apenas)
        n = len(X[:, 0]) # Numero de observações
    else:
        k = len(X[0,:] ) + 1 # Graus de liberade de regressão (graus de inclinação + intercepto)
        n = len(X[:, 0]) # Numero de observações
    
    params = np.append(plm.intercept_,plm.coef_)
    std_error = stderror_plm(X, y, plm)
    z_values = params/std_error # Coeficientes do teste t para os parâmetros do modelo
    p_values =[2*(1-stats.norm.cdf(abs(i))) for i in z_values]
    
    
    return z_values, p_values

def OtherTests_plm(X, y, plm):
    from scipy.special import gamma
    if isinstance(X, pd.DataFrame):
        if len(X.columns)==1:
            X = X.values.reshape(-1,1)
            y = y.valueslm.score(X, y)
        else:
            X = X.values
            y = y.values
    
    k = len(X[0,:] ) # Numero de variáveis
    n = len(X[:, 0]) # Numero de observações
    
    loglike = np.sum(np.ravel(y) * np.log(plm.predict(X)) - plm.predict(X) - np.ravel(np.log(gamma(y + 1))))
    
    X_null = np.ones_like(y).reshape(-1, 1)
    plm_null = PoissonRegressor(fit_intercept=True)
    plm_null = plm_null.fit(X_null, y)
    
    loglike0 = np.sum(np.ravel(y) * np.log(plm_null.predict(X_null)) - plm_null.predict(X_null) - np.ravel(np.log(gamma(y + 1))))

    chi2_value = -2*(loglike0 - loglike)
    p_value = 1-stats.chi2.cdf(chi2_value, k)
    
    AIC = -2*loglike + 2*(k+1)
    BIC = -2*loglike + (k + 1)*np.log(n)
    pR2_McFadden = chi2_value/(-2*loglike0)
    pR2_Cragg = (1 - (np.exp(loglike0)/np.exp(loglike))**(2/n))/(1 - np.exp(loglike0)**(2/n))
    
    values = [int(n),n -k, len(X[0,:] ),  chi2_value, p_value, loglike, AIC, BIC, pR2_McFadden, pR2_Cragg]
    index = ['No. Observations', 'Df Residuals', 'Df Model','Chi2', 'p-value (Chi2)', 'Log-Likelihood', 'AIC', 'BIC', 'pseudo-R2 (McFadden)', 'pseudo-R2 (Cragg)']
    df = pd.DataFrame({'Values': values}, index = index)
    
    return df

df_params, df_other = getParams_plm(X, y, plm)
```

## 1.3. Teste de Cameron-Trivedi

```python
# Todas as variáveis preditoras se mostraram estatisticamente
#diferentes de zero, considerando-se um nível de significância de 5%,
#ceteris paribus. Porém, já se pode afirmar que a estimação Poisson é a mais
#adequada?

################################################################################
#            TESTE DE SUPERDISPERSÃO DE CAMERON E TRIVEDI (1990)               #
################################################################################
#CAMERON, A. C.; TRIVEDI, P. K. Regression-based tests for overdispersion in
#the Poisson model. Journal of Econometrics, v. 46, n. 3, p. 347-364, 1990.

#1º Passo: estimar um modelo Poisson;
#2º Passo: criar uma nova variável (Y*) utilizando os fitted values do modelo
#Poisson estimado anteriormente;
#3º Passo: estimar um modelo auxiliar OLS, com a variável Y* como variável
#dependente, os fitted values do modelo Poisson como única variável preditora e 
#sem o intercepto;
#4º Passo: Observar a significância do parâmetro beta.


#Adicionando os fitted values do modelo Poisson (lambda_poisson) ao dataframe:
df['lambda_poisson'] = modelo_poisson.fittedvalues


#Criando a nova variável Y*:
df['ystar'] = (((df['violations']
                            -df['lambda_poisson'])**2)
                          -df['violations'])/df['lambda_poisson']
df


#Adicionando os fitted values do modelo Poisson (lambda_poisson) ao dataframe:
df['lambda_poisson'] = modelo_poisson.fittedvalues
df

#Criando a nova variável Y*:
df['ystar'] = (((df['violations']
                            -df['lambda_poisson'])**2)
                          -df['violations'])/df['lambda_poisson']
df



```

```python
#Estimando o modelo auxiliar OLS, sem o intercepto:
modelo_auxiliar = smf.ols(formula='ystar ~ 0 + lambda_poisson',
                          data=df).fit()

#Parâmetros do 'modelo_auxiliar'
modelo_auxiliar.summary()

#Caso o p-value do parâmetro do lambda_poisson seja maior que 0.05,
#verifica-se a existência de equidispersão nos dados.
#Caso contrário, diagnostica-se a existência de superdispersão nos dados, fato
#que favorecerá a estimação de um modelo binomial negativo, como ocorre nesse
#caso.
```

```python
# Função 'overdisp'
# Instalação e carregamento da função 'overdisp' do pacote 'statstests.tests'
# Autores do pacote: Luiz Paulo Fávero e Helder Prado Santos
# https://stats-tests.github.io/statstests/
# pip install statstests
from statstests.tests import overdisp

#Elaboração direta do teste de superdispersão
overdisp(modelo_poisson, df)
```

### - statsmodels.discrete.discrete_model.NegativeBinomial

```python
# Esta implementação já calcula o alpha, mas nos dá valores de z diferentes.
import statsmodels.discrete

X = df[['staff', 'post', 'corruption']]
y = df[['violations']]
X_with_constant = sm.tools.tools.add_constant(X, prepend= True)

modelo_bneg = statsmodels.discrete.discrete_model.NegativeBinomial(endog=y, exog=X_with_constant).fit()
modelo_bneg.summary()

```

```python
def getParams_bneg(X, y, modelo_bneg):

    index = ['(Intercept)'] + list(X.columns.values)
    params = modelo_bneg.params.drop('alpha').values
    std_error = stderror_bneg(X, y, modelo_bneg)
    z_values, p_values = z_test_bneg(X, y, modelo_bneg)
    df = pd.DataFrame({'Coefficients': params, "Standard Errors": std_error, "z-statistics": z_values, 'p-values': p_values}, index = index)
    
    #other_df = OtherTests_plm(X, y, plm)
    #return df, other_df
    return df

def stderror_bneg(X, y, modelo_bneg):
    if isinstance(X, pd.DataFrame):
        if len(X.columns)==1:
            X = X.values.reshape(-1,1)
            y = y.values
        else:
            X = X.values
            y = y.values
    
    if isinstance(X, pd.DataFrame):
        if len(X.columns)==1:
            X = X.values.reshape(-1,1)
            y = y.values
        else:
            X = X.values
            y = y.values

    params = modelo_bneg.params.drop('alpha').values
    alpha = modelo_bneg.params['alpha']
    X_matrix = np.insert(X, 0, 1, axis = 1)
    W_matrix = np.diag(np.exp(modelo_bneg.fittedvalues.values)/(1 + alpha*np.exp(modelo_bneg.fittedvalues.values)))
    C_matrix = np.linalg.inv(np.matmul(X_matrix.T,np.matmul(W_matrix, X_matrix)))#
    C_diag = np.diag(C_matrix)
    std_error = np.sqrt(C_diag)
    
    return std_error


def z_test_bneg(X, y, modelo_bneg):
    
    if isinstance(X, pd.DataFrame):
        if len(X.columns)==1:
            X = X.values.reshape(-1,1)
            y = y.values
        else:
            X = X.values
            y = y.values

    params = modelo_bneg.params.drop('alpha').values
    std_error = stderror_bneg(X, y, modelo_bneg)
    z_values = params/std_error # Coeficientes do teste \ para os parâmetros do modelo
    p_values =[2*(1-stats.norm.cdf(abs(i))) for i in z_values]
    
    
    return z_values, p_values
```

## 2.2. Comparação entre modelos

### - Comparação dos parâmetros

```python
# Comparando os modelos Poisson e binomial negativo

summary_col([modelo_poisson, modelo_bneg], 
            model_names=["Poisson","BNeg"],
            stars=True,
            info_dict = {
                'N':lambda x: "{0:d}".format(int(x.nobs)),
                'Log-lik':lambda x: "{:.2f}".format(x.llf),
                'Pseudo-R2':lambda x: "{:.4f}".format(x.pseudo_rsquared()),
        })

```

```python
# likelihood ratio test para comparação de LL's entre modelos

#Definição da função 'lrtest'
def lrtest(modelos):
    modelo_1 = modelos[0]
    llk_1 = modelo_1.llnull
    llk_2 = modelo_1.llf
    
    if len(modelos)>1:
        llk_1 = modelo_1.llf
        llk_2 = modelos[1].llf
    LR_statistic = -2*(llk_1-llk_2)
    p_val = stats.chi2.sf(LR_statistic, 1)
    return round(LR_statistic,2), round(p_val,2)

lrtest([modelo_poisson, modelo_bneg])
```

# 3. Modelo Zero Inflated Poisson

## 3.1. Estimação do modelo

```python
#Definição da variável dependente (voltando ao dataset 'df_corruption')
y = df.violations

#Definição das variáveis preditoras que entrarão no componente de contagem
x1 = df[['staff','post','corruption']]
X1 = sm.add_constant(x1)

#Definição das variáveis preditoras que entrarão no componente logit (inflate)
x2 = df[['corruption']]
X2 = sm.add_constant(x2)



#Estimação do modelo ZIP pela função 'ZeroInflatedPoisson' do pacote
#'Statsmodels'

#Estimação do modelo ZIP
#O argumento 'exog_infl' corresponde às variáveis que entram no componente
#logit (inflate)
modelo_zip = sm.ZeroInflatedPoisson(y, X1, exog_infl=X2,
                                    inflation='logit').fit()

#Parâmetros do modelo
modelo_zip.summary()
```

```python
#Gráfico para comparar os valores previstos x valores reais de
#'violations' pelo modelo ZIP

zip_predictions = modelo_zip.predict(X1, exog_infl=X2)
predicted_counts = np.round(zip_predictions)
actual_counts = df['violations']

plt.figure(figsize=(15,10))
plt.plot(df.index, predicted_counts, 'go-',
         color='orange')
plt.plot(df.index, actual_counts, 'go-',
         color='#440154FF')
plt.xlabel('Observação', fontsize=20)
plt.ylabel('Violações de Trânsito', fontsize=20)
plt.legend(['Valores Previstos pelo ZIP', 'Valores Reais no Dataset'],
           fontsize=20)
plt.show()
```

## 3.2. Comparação entre modelos

### - Comparação dos parâmetros

```python
#Comparando os modelos Poisson e ZIP

summary_col([modelo_poisson, modelo_zip], 
            model_names=["Poisson","ZIP"],
            stars=True,
            info_dict = {
                'N':lambda x: "{0:d}".format(int(x.nobs)),
                'Log-lik':lambda x: "{:.2f}".format(x.llf),
                'Pseudo-R2':lambda x: "{:.4f}".format(x.pseudo_rsquared()),
        })
```

```python
#likelihood ratio test para comparação de LL's entre modelos

#Definição da função 'lrtest'
def lrtest(modelos):
    modelo_1 = modelos[0]
    llk_1 = modelo_1.llnull
    llk_2 = modelo_1.llf
    
    if len(modelos)>1:
        llk_1 = modelo_1.llf
        llk_2 = modelos[1].llf
    LR_statistic = -2*(llk_1-llk_2)
    p_val = stats.chi2.sf(LR_statistic, 1)
    return round(LR_statistic,2), round(p_val,2)

lrtest([modelo_poisson, modelo_zip])
```

## 4.1. Estimação do modelo

```python
# 
##############################################################################
#        ESTIMAÇÃO DO MODELO ZERO-INFLATED BINOMIAL NEGATIVO (ZINB)          #
##############################################################################

#Definição da variável dependente (voltando ao dataset 'df_corruption')
y = df.violations

#Definição das variáveis preditoras que entrarão no componente de contagem
x1 = df[['staff','post','corruption']]
X1 = sm.add_constant(x1)

#Definição das variáveis preditoras que entrarão no componente logit (inflate)
x2 = df[['corruption']]
X2 = sm.add_constant(x2)


#Estimação do modelo ZINB pela função 'ZeroInflatedNegativeBinomialP' do
#pacote 'statsmodels.discrete.count_model'

from statsmodels.discrete.count_model import ZeroInflatedNegativeBinomialP

#Estimação do modelo ZINB
#O argumento 'exog_infl' corresponde às variáveis que entram no componente
#logit (inflate)
modelo_zinb = ZeroInflatedNegativeBinomialP(y, X1, exog_infl=X2,
                                            inflation='logit').fit()

#Parâmetros do modelo
modelo_zinb.summary()

#O parâmetro 'alpha' apresentado é o inverso do parâmetro 'theta', ou seja,
#o inverso do parâmetro de forma da distribuição Poisson-Gama.
#Como 'alpha' (e da mesma forma 'theta') é estatisticamente diferente de
#zero, podemos afirmar que há superdispersão nos dados (outra forma de
#verificar o fenômeno da superdispersão!)
```

```python
# Gráfico para comparar os valores previstos x valores reais de
#'violations' pelo modelo ZINB

zinb_predictions = modelo_zinb.predict(X1, exog_infl=X2)
predicted_counts = np.round(zinb_predictions)
actual_counts = df['violations']

plt.figure(figsize=(15,10))
plt.plot(df.index, predicted_counts, 'go-',
         color='orange')
plt.plot(df.index, actual_counts, 'go-',
         color='#440154FF')
plt.xlabel('Observação', fontsize=20)
plt.ylabel('Violações de Trânsito', fontsize=20)
plt.legend(['Valores Previstos pelo ZINB', 'Valores Reais no Dataset'],
           fontsize=20)
plt.show()
```
