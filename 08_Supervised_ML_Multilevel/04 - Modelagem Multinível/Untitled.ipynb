```python
import numpy as np
from scipy.special import gamma, factorial
import matplotlib.pyplot as plt
def omega(V, N, E):
    const = (2 * np.pi)**(3*N/2) / gamma(3*N/2 )
    return V**N * const * (2 * E)**(3*N/2 - 1)

def Omega(V, N, E):
    const = np.pi**(3*N/2) / gamma(3*N/2 + 1)
    return V**N * const * (2 * E)**(3*N/2 )

N_array = np.arange(50, 1000)
fig, ax = plt.subplots()

# Plot the data on the axes
ax.plot(N_array, omega(1, 100, N_array))
#ax.plot(N_array, Omega(1, N_array, 10))

# Add labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Sine Wave')

# Show the plot
plt.show()
```
