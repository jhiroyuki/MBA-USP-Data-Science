### - Sklearn

```python
#Estimando o modelo usando sklearn
from sklearn.linear_model import LinearRegression


X = df[['distancia']]
y = df[['tempo']]
lm = LinearRegression()
lm = lm.fit(X.values.reshape(-1,1), y.values)

# Não há tabela de resultados como o statsmodel. 

# Vamos criar uma função para tirar os resultados dos coeficientes e realizar os testes
from scipy import stats
def getParams_lm(X, y, lm):

    index = ['(Intercept)'] + list(X.columns.values)
    params = np.append(lm.intercept_,lm.coef_)
    df = pd.DataFrame({'Coefficients': params}, index = index)
    std_error = stderror_lm(X, y, lm)
    t_values, p_values = t_test_lm(X, y, lm)
    df = pd.DataFrame({'Coefficients': params, "Standard Errors": std_error,
                          "t-statistics": t_values, 'p-values': p_values}, index = index)
    other_df = OtherTests_lm(X, y, lm)
    return df, other_df


def stderror_lm(X, y, lm):
    if isinstance(X, pd.DataFrame):
        if len(X.columns)==1:
            X = X.values.reshape(-1,1)
            y = y.values
        else:
            X = X.values
            y = y.values
    
    if lm.intercept_ ==0:
        k = len(X[0,:] ) # Graus de liberade de regressão (graus de inclinação apenas)
        n = len(X[:, 0]) # Numero de observações
    else:
        k = len(X[0,:] ) + 1 # Graus de liberade de regressão (graus de inclinação + intercepto)
        n = len(X[:, 0]) # Numero de observações

    params = np.append(lm.intercept_,lm.coef_)
    predictions = lm.predict(X) 
    
    X_matrix = np.insert(X, 0, 1, axis = 1) # Matriz de design
    MSE = sum((y-predictions)**2)/(n - k ) # Mean Squared Error
    C_matrix = MSE*(np.linalg.inv(np.dot(X_matrix.T,X_matrix))) # Variance-covariance matrix
    C_diag = np.diag(C_matrix)
    std_error = np.sqrt(C_diag) # Erros padrões
    return std_error

# Teste t
def t_test_lm(X, y, lm):
    
    if isinstance(X, pd.DataFrame):
        if len(X.columns)==1:
            X = X.values.reshape(-1,1)
            y = y.values
        else:
            X = X.values
            y = y.values

    
    if lm.intercept_ ==0:
        k = len(X[0,:] ) # Graus de liberade de regressão (graus de inclinação apenas)
        n = len(X[:, 0]) # Numero de observações
    else:
        k = len(X[0,:] ) + 1 # Graus de liberade de regressão (graus de inclinação + intercepto)
        n = len(X[:, 0]) # Numero de observações
    
    params = np.append(lm.intercept_,lm.coef_)
    std_error = stderror_lm(X, y, lm)
    t_values = params/std_error # Coeficientes do teste t para os parâmetros do modelo
    p_values =[2*(1-stats.t.cdf(np.abs(i),(n - k))) for i in t_values] 
    
    
    return t_values, p_values

def OtherTests_lm(X, y, lm):
    
    if isinstance(X, pd.DataFrame):
        if len(X.columns)==1:
            X = X.values.reshape(-1,1)
            y = y.values
        else:
            X = X.values
            y = y.values
    
    if lm.intercept_ ==0:
        k = len(X[0,:] ) # Graus de liberade de regressão (graus de inclinação apenas)
        n = len(X[:, 0]) # Numero de observações
    else:
        k = len(X[0,:] ) + 1 # Graus de liberade de regressão (graus de inclinação + intercepto)
        n = len(X[:, 0]) # Numero de observações
        
    Rsqr = lm.score(X, y) # R-squared
    Rsqr_adjusted = 1 - (n-1)/(n-k) * (1 - Rsqr) # Adjusted R-squared

    # Teste F
    F_value = (Rsqr/(k -1))/((1- Rsqr)/(n - k )) 
    p_value = 1-stats.f.cdf(F_value, k-1, n-k)
    
    
    var_residues = np.sum((y-lm.predict(X) )**2)/(n) # Variância dos resíduos
    loglike = -n/2 * (np.log(2*np.pi*var_residues ) + 1) # Log-likelihood
    
    
    AIC = -2*loglike + 2*k 
    BIC = -2*loglike + k*np.log(n)  
    
    
    values = [int(n),n -k, len(X[0,:] ),   Rsqr, Rsqr_adjusted, F_value, p_value, loglike, AIC, BIC]
    index = ['No. Observations', 'Df Residuals', 'Df Model','R-squared', 'Adj. R-squared','F-statistic', 'p-value (F-statistic)', 'Log-Likelihood', 'AIC', 'BIC']
    df = pd.DataFrame({'Values': values}, index = index)

    return df
    

df_params, df_other = getParams_lm(X, y, lm)
```

### - Sklearn

```python
# Intervalo de confiança usando sklearn
def confint_lm(X, y, lm, significance =0.05):
    index = ['(Intercept)'] + list(X.columns.values)
    
    if isinstance(X, pd.DataFrame):
        if len(X.columns)==1:
            X = X.values.reshape(-1,1)
            y = y.values
        else:
            X = X.values
            y = y.values
    
    if lm.intercept_ ==0:
        k = len(X[0,:] ) # Graus de liberade de regressão (graus de inclinação apenas)
        n = len(X[:, 0]) # Numero de observações
    else:
        k = len(X[0,:] ) + 1 # Graus de liberade de regressão (graus de inclinação + intercepto)
        n = len(X[:, 0]) # Numero de observações

    params = np.append(lm.intercept_,lm.coef_)
    
    stderror = stderror_lm(X, y, lm) 
    t_critical = stats.t.ppf(1- significance/2, n - k) # Valor em x da distriuição em t que corresponde ao valor de significancia
    
    params_correction = t_critical*stderror
    params_low = params - params_correction
    params_high = params + params_correction

    df = pd.DataFrame({'Coefficients': params,str(significance/2*100) +'%': params_low,str((1-significance/2)*100) +'%': params_high }, index= index)
    return df

confint_lm(X, y, lm, significance =0.05)
```

# 4. Box-Cox

### - 2º passo: Shapiro-Francia

```python
# In[ ]: Teste de verificação da aderência dos resíduos à normalidade

# Teste de Shapiro-Wilk (n < 30)
#from scipy.stats import shapiro
#shapiro(modelo_linear.resid)

# Teste de Shapiro-Francia (n >= 30)
# Instalação e carregamento da função 'shapiroFrancia' do pacote
#'sfrancia'
# Autores: Luiz Paulo Fávero e Helder Prado Santos
#pip install sfrancia==1.0.8
from sfrancia import shapiroFrancia
shapiroFrancia(modelo_linear.resid)

# Se o p-value é menor que 0.05 então os resíduos não são aderentes a normalidade
```

```python
# Fazendo manualmente

def shapiroFrancia(x):
    x = np.sort(x)
    n = len(x)
    y = stats.norm.ppf(ppoints(n,a = 3/8))
    W = np.corrcoef(x, y)**2
    u = np.log(n)
    v = np.log(u)
    mu = -1.2725 + 1.0521 * (v - u)
    sig = 1.0308 - 0.26758 * (v + 2/u)
    z = (np.log(1 - W) - mu)/sig
    pval = stats.norm.sf(z)

    SF_df = pd.DataFrame()
    SF_df['W'] = [removerDiagonal(W)[0][0]]
    SF_df['p-value'] =  [removerDiagonal(pval)[0][0]]

    print('\n######## SHAPIRO FRANCIA ##########')
    print(SF_df)
    
def ppoints(n, a):   
    try:
        n = float(len(n))
    except TypeError:
        n = float(n)
    return (np.arange(n) + 1 - a)/(n + 1 - 2*a)

def removerDiagonal(x):
    x_sem_diagonais = np.ndarray.flatten(x)
    x_sem_diagonais = np.delete(x_sem_diagonais, range(0, len(x_sem_diagonais), len(x) + 1), 0)
    x_sem_diagonais = x_sem_diagonais.reshape(len(x), len(x) - 1)
    return x_sem_diagonais     



shapiroFrancia(modelo_linear.resid)
```

## 7.3. Estimação de modelo e diagnóstico

### - 1º Passo: Regressão linear

```python
# Estimação do modelo de regressão e diagnóstico de heterocedasticidade

# Estimando o modelo
modelo = sm.OLS.from_formula('saeb ~ rendimento', df).fit()

# Parâmetros do modelo
modelo.summary()
```

### - 2º Passo: Breusch-Pagan

```python
# Função para o teste de Breusch-Pagan para a elaboração
# de diagnóstico de heterocedasticidade

# Criação da função 'breusch_pagan_test'

from scipy import stats

def breusch_pagan_test(modelo):

    df = pd.DataFrame({'yhat':modelo.fittedvalues,
                       'resid':modelo.resid})
   
    df['up'] = (np.square(df.resid))/np.sum(((np.square(df.resid))/df.shape[0]))
   
    modelo_aux = sm.OLS.from_formula('up ~ yhat', df).fit()
   
    anova_table = sm.stats.anova_lm(modelo_aux, typ=2)
   
    anova_table['sum_sq'] = anova_table['sum_sq']/2
    
    chisq = anova_table['sum_sq'].iloc[0]
   
    p_value = stats.chi2.pdf(chisq, 1)*2
    
    print(f"chisq: {chisq}")
    
    print(f"p-value: {p_value}")
    
    return chisq, p_value
```

```python
# Teste de Breusch-Pagan propriamente dito

breusch_pagan_test(modelo)
#Presença de heterocedasticidade -> omissão de variável(is) explicativa(s)
#relevante(s)

#H0 do teste: ausência de heterocedasticidade.
#H1 do teste: heterocedasticidade, ou seja, correlação entre resíduos e
#uma ou mais variáveis explicativas, o que indica omissão de
#variável relevante!
```

### - 3º Passo: Regressão linear com adição de novas variáveis

### - 4º Passo: Breusch-Pagan

```python
# Função para o teste de Breusch-Pagan para a elaboração
# de diagnóstico de heterocedasticidade

# Criação da função 'breusch_pagan_test'

#from scipy import stats

def breusch_pagan_test(modelo):

    df = pd.DataFrame({'yhat':modelo.fittedvalues,
                       'resid':modelo.resid})
   
    df['up'] = (np.square(df.resid))/np.sum(((np.square(df.resid))/df.shape[0]))
   
    modelo_aux = sm.OLS.from_formula('up ~ yhat', df).fit()
   
    anova_table = sm.stats.anova_lm(modelo_aux, typ=2)
   
    anova_table['sum_sq'] = anova_table['sum_sq']/2
    
    chisq = anova_table['sum_sq'].iloc[0]
   
    p_value = stats.chi2.pdf(chisq, 1)*2
    
    print(f"chisq: {chisq}")
    
    print(f"p-value: {p_value}")
    
    return chisq, p_value


# Teste de Breusch-Pagan propriamente dito

breusch_pagan_test(modelo_step)
#Presença de heterocedasticidade -> omissão de variável(is) explicativa(s)
#relevante(s)

#H0 do teste: ausência de heterocedasticidade.
#H1 do teste: heterocedasticidade, ou seja, correlação entre resíduos e
#uma ou mais variáveis explicativas, o que indica omissão de
#variável relevante!

```
