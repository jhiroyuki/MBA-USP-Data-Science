## 2.2. Componentes tendência, sazonal e erros

```python
# Criando nova série temporal

# Definir a sequência de valores
valores = [10, 14, 8, 25, 16, 22, 14, 35, 15, 27, 18, 40, 28, 40, 25, 65]

# Criar o índice temporal
start = pd.Timestamp('2019-01-01')
end = pd.Timestamp('2022-12-31')
idx = pd.date_range(start, end, freq='3M')
idx

# Criar o DataFrame da série temporal
ts = pd.DataFrame(valores, index=idx, columns=['Valor'])

ts
```

### - Modelo aditivo

```python
# Decomposição pelo modelo ADITIVO
decomposition_additive = sm.tsa.seasonal_decompose(ts, model='additive')
```

```python

# Plotar os resultados
fig, axes = plt.subplots(4, 1, figsize=(8, 10))

axes[0].plot(ts, label='Original')
axes[0].legend()

axes[1].plot(decomposition_additive.trend, label='Tendência')
axes[1].legend()

axes[2].plot(decomposition_additive.seasonal, label='Sazonalidade')
axes[2].legend()

axes[3].plot(decomposition_additive.resid, label='Resíduo')
axes[3].legend()

plt.tight_layout()
plt.show()
```

```python
# Note que a tendencia é a média movel de 4 períodos.

ts_media_movel = ts.rolling(window=4).mean()
ts_tendencia = (ts_media_movel + ts_media_movel.shift(1)) / 2
ts_tendencia = ts_tendencia.shift(-2)

pd.DataFrame({'manual': np.ravel(ts_tendencia.values), 'statsmodel': decomposition_additive.trend.values}, index = ts_tendencia.index)
```

```python
# Para calcular a sazonalidade, primeiro subtraimos as tendências das observações originais 

ts_subtract_trend = ts - ts_tendencia

# Fazemos essa matriz, onde cada coluna é um dos quadrimestres de cada ano
ts_subtract_matriz = ts_subtract_trend.values.reshape((4,4))

# Em seguida tiramos a média de cada quadrimestre
ts_subtract_mean = ts_subtract_trend.groupby(ts_subtract_trend.index.quarter)['Valor'].mean()

# Média total das médias
total_mean  = ts_subtract_mean.mean()

# Os elementos da sazonalidade são os ajustes das médias
seasonal_elements = ts_subtract_mean - total_mean
ts_season = np.tile(seasonal_elements, 4)
# Construindo o dataframe 
pd.DataFrame({'manual': ts_season, 'statsmodel': decomposition_additive.seasonal.values}, index = ts_tendencia.index)
```

```python
# O ciclo + erro é encontrado pela subtração das observações + tendencia da sasonalidade

ts_resid = np.ravel(ts_subtract_trend.values) - ts_season

# Construindo o dataframe 
pd.DataFrame({'manual': ts_resid, 'statsmodel': decomposition_additive.resid.values}, index = ts_tendencia.index)
```

### - Modelo multiplicativo

```python
# Decomposição pelo modelo MULTIPLICATIVO
decomposition_multiplicative = sm.tsa.seasonal_decompose(ts, model='multiplicative')

# Neste caso, a forma de encontrar as componentes é a mesma, substituindo a subtração por divisão.
```

```python

# Plotar os resultados
fig, axes = plt.subplots(4, 1, figsize=(8, 10))

axes[0].plot(ts, label='Original')
axes[0].legend()

axes[1].plot(decomposition_multiplicative.trend, label='Tendência')
axes[1].legend()

axes[2].plot(decomposition_multiplicative.seasonal, label='Sazonalidade')
axes[2].legend()

axes[3].plot(decomposition_multiplicative.resid, label='Resíduo')
axes[3].legend()

plt.tight_layout()
plt.show()
```

## 3.5. Exemplo para comparação de modelos

```python
# Importando os dados
df = pd.read_csv('datasets/ambev.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# Definir a coluna de data como o índice (opcional)
df.set_index('Data', inplace=True)
df.head()
```

```python
# Criar time series para a variável 'Fechamento'
ts = df[['receita_ac']]
ts
```

```python
ts.info()
```

```python
# Mudar tipo de index
ts.index = pd.to_datetime(ts.index)
```

```python
ts.info()
```

```python
# Separar a base de dados em uma janela para criar o modelo e outra para prever
start_date = '2000-01-01'
end_date = '2020-12-31'
ts_train = ts[start_date:end_date]

# Separar a base de dados em uma janela para criar o modelo e outra para prever
start_date = '2021-01-01'
end_date = '2022-12-31'
ts_teste = ts[start_date:end_date]
```

```python
# Gráfico de série temporal

fig, ax = plt.subplots(figsize=(10, 6))

# Plotar a série temporal
ax.plot(ts_train, color='black', label='Treino')
ax.plot(ts_teste, color='red', label='Teste')

# Personalizar o gráfico
ax.set_xlabel('Trimestres - jan/2000 a dez/2022"')
ax.set_ylabel('R$ mil')
ax.set_title('Faturamento Trimestral - Acumulado da AMBEV SA - 1T/2000 ao 4T/2022')
ax.xaxis.set_tick_params(rotation=90, labelsize=8)

#ax.grid(color='grey', linestyle='-', linewidth=0.5)
ax.legend().remove()

# Exibir o gráfico
plt.show()
```

```python
# Suavização Exponencial Simples
from statsmodels.tsa.holtwinters import SimpleExpSmoothing

# Criando o modelo de suavização exponencial
model_ses = SimpleExpSmoothing(ts_train)
model_ses = model_ses.fit()
```

```python
# Fazer a previsão para 8 passos à frente
forecast_ses = model_ses.forecast(steps=8)
forecast_ses # Valores previstos
```

```python
# Mean Absolute Percentage Error (MAPE)
from sklearn.metrics import mean_absolute_percentage_error

mape_ses = mean_absolute_percentage_error(ts_teste, forecast_ses)
mape_ses
```

### - Testes de estacionaridade

#### Dickey-Fuller

```python
from statsmodels.tsa.stattools import adfuller

# Teste de Dickey-Fuller
# H0: A série Não é Estacionária
# H1: A série é Estacionária


# Teste para AR(1)

# Realizar o teste de Dickey-Fuller
result = adfuller(arima100_data)

# Interpretar o resultado do teste
adf_statistic = result[0]
p_value = result[1]
critical_values = result[4]

print(f"Resultado do teste de Dickey-Fuller:")
print(f"Estatística ADF: {adf_statistic}")
print(f"Valor-p: {p_value}")
print("Valores críticos:")
for key, value in critical_values.items():
    print(f"  {key}: {value}")

# Verificar se a série é estacionária com base no valor-p
if p_value <= 0.05:
    print("A série é estacionária (rejeita-se a hipótese nula).")
else:
    print("A série não é estacionária (não se rejeita a hipótese nula).")
```

### - ACF e PACF (Funções de autocorrelação)

```python
from statsmodels.tsa.stattools import acf, pacf
# ACF - Função de autocorrelação
# Mostra as
# Linhas mostram significância (intervalo de confiança) autocorrelações em uma série temporal
# A 1ª autocorrelação é igual a 1. Cada traço do gráfico mostra uma defasagem e uma correlação (autocorrelação).

# PACF - Função de autocorrelação parcial
# Mede a autocorrelação não entre lags mas entre diferentes intervalos.

# Calcular a ACF e PACF para AR(1)
lag_max = 5
acf_values, confint = acf(arima100_data, fft=False, alpha=0.05)
pacf_values, confint_pacf = pacf(arima100_data, nlags=lag_max, alpha=0.05)

# Plotar a ACF
plt.figure(figsize=(10, 6))
plt.stem( acf_values, linefmt='b-', markerfmt='bo', basefmt='r-')
plt.title('Função de Autocorrelação (ACF)')
plt.xlabel('Lag')
plt.ylabel('ACF')
plt.grid(True)
plt.show()

# Plotar a PACF
plt.figure(figsize=(10, 6))
plt.stem(range(lag_max+1), pacf_values, linefmt='b-', markerfmt='bo', basefmt='r-')
plt.title('Função de Autocorrelação Parcial (PACF)')
plt.xlabel('Lag')
plt.ylabel('PACF')
plt.grid(True)
plt.show()

# Padrão típico da ACF para AR(p)
# Decai exponencialmente para zero ou com padrão de onda senoidal amortecida, ou ambos.

# Padrão típico da PACF para AR(p)
# Valores significativos, ou seja, não nulos, até a defasagem p.
```
